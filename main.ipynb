```python
import requests
import pandas as pd

# URL of the file
url = "https://data.humdata.org/dataset/dc663585-4b6f-46ae-a6d6-b2f3e4ea32b5/resource/3b1ff071-6b01-4998-aa62-2f3efb5d4888/download/wfp_food_prices_ind.csv"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Write the content to a local file
    with open('wfp_food_prices_ind.csv', 'wb') as file:
        file.write(response.content)
    
    # Load the CSV file into a pandas DataFrame
    df = pd.read_csv('wfp_food_prices_ind.csv')
    
    # Display the first few rows of the DataFrame
    print(df.head())
else:
    print(f"Failed to download file. Status code: {response.status_code}")
```
